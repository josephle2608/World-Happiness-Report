<h1 align=center><font size = 5>World Happiness Report</font></h1>

Every year there is an interesting survey about the state of worldwide bliss which is World Happiness Report. The report shows rankings of countries in the world from most happy countries  to least happy countries. There are many features that affect to the rankings include financial matters,  overview investigation, national insights, wellbeing, and more. For the project, the dataset is collected from kaggle.com website. This is the link to the dataset:

[**World Happiness Report**](https://www.kaggle.com/mathurinache/world-happiness-report)

There are datasets for every year but the dataset 2019 is used for the project. There are 156 rows which represent 156 countries or regions in the dataset and 9 variables shown as follows:
- Overall rank: rankings from most happy countries to the least countries
- Country or region: names of countries or region
- Score: the overall score of happiness for the country
- GDP per capita: GDP as an economic index of a country
- Social support: how family and friends give support to each other 
- Healthy life expectancy: the expected number of life years spent in good health
- Freedom to make life choice: index of satisfy with ones’ freedom to choose what they do with their lives
- Generosity: the quality of being kind and generous
- Perceptions of corruption: index shown as perceived levels of public sector corruption

## Data Exploration
In the dataset, variables including Country or region, Score, Overall rank, GDP per capita and Healthy Life Expectancy are collected to be analyzed. I’d like to know how economy and health care can affect the happiness of a country so GDP per capita and Healthy life expectancy variables were chosen. Score and Overall rank are considered as target variables. Country or region variable is used to represent names of countries.


```python
ax = sns.regplot(x='GDP per capita', y='Score', data=df)
ax.axes.set_xlim(0,)
ax.axes.set_ylim(2,)
```


The first basic graph is scatter plot. The graph showed the relationship between Score and GDP per capita. It indicated the countries with higher GDP per capita have higher happiness scores. 


```python
life_expectancy = df.iloc[[0,1,2,76,77,78,153,154,155],[0,1,5]]
life_expectancy
```

```python
plt.figure(figsize=(15,10))
ax = sns.barplot(x = 'Overall rank', y = 'Healthy life expectancy', hue='Country or region',data=life_expectancy,dodge=False)
```

The second basic graph is bar plot. The graph showed the comparison of healthy life expectancy between top 3 happiness countries with three countries at middle of rankings and three least happiness countries. It indicated that the healthy life expectancy lowered as rankings were further down.

## The Relationship Between The Selected Variables With Other Variables

```python
df_1 = df.drop(['Country or region'], axis=1)
ax = sns.heatmap(df_1.corr(), cmap="YlGnBu", annot=True)
```

The heatmap showed correlations between variables. Based on the graph, we can see relationships between Overall rank, Score variables and other variables.

```python
sns.pairplot(df,
             x_vars=["Social support", "Freedom to make life choices", "Generosity", "Perceptions of corruption"],
            y_vars=["GDP per capita", "Healthy life expectancy"]
            )
```

The scatter plots showed relationships between GDP per capita, Healthy life expectancy variables and other variables.

## Discussion
Based on the heatmap, Overall rank and Score variables have strong correlation with GDP per capita, Social Support, Healthy life expectancy, and Freedom to make life choices variables. They showed low correlation with Perception of corruption variable and nearly no correlation with Generosity variable.

Based on the scatter plots, GDP per capita and Healthy life expectancy variables showed strong correlation with Social support. They have low correlation with Freedom to make life choices and Perception of corruption variables. They nearly don’t have correlation with Generosity variable.
